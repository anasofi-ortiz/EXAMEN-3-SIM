```python
import matplotlib.pyplot as plt    
import numpy as np                      
from scipy.integrate import odeint     
%matplotlib inline  
import numpy as np, matplotlib.pyplot as plt, matplotlib.font_manager as fm, os
from mpl_toolkits.mplot3d.axes3d import Axes3D
import matplotlib as mpl
```

PRIMER PUNTO

```python
# Definimos los parámetros
a = 1.0                                     
b = 1.0                                        
                                       

# Función f(t,y) que vamos a integrar
def armonico(y, t, a, b):
    #y1, y2 = y
    y1 = y[0]
    y2 = y[1]
    return [a * y2, -b * y1]

y0 = [5, 5]                                 
                                             

tt = np.linspace(0, 50, 500)                    

yy = odeint(armonico, y0, tt, args=(a, b))
```

```python
y1 = yy[:,0]
y2 = yy[:,1]
```

```python
label_size = 14
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size 
```

```python
plt.figure(figsize=(8,6))
plt.plot(tt, y1, 'b', label='tiempo')
plt.plot(tt, y2, 'r', label='tiempo')
plt.legend(loc='best')
plt.xlabel('tiempo')
plt.grid()
plt.show()
```

```python

plt.figure(figsize=(8,8))
plt.scatter(y1, y2, c=y2, cmap='viridis')
plt.xlabel('tiempo')
plt.ylabel('tiempo')
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(12, 9))
ax = fig.gca(projection='3d')
ax.xaxis.set_pane_color((1,1,1,1))
ax.yaxis.set_pane_color((1,1,1,1))
ax.zaxis.set_pane_color((1,1,1,1))
ax.plot(tt, y1, y2, color='g', alpha=0.7, linewidth=1)

ax.set_title('El Amor');
ax.set_xlabel('tiempo')
ax.set_ylabel('Romeo')
ax.set_zlabel('Julieta');

plt.show()
```

Una vez resuelta la ecuacion diferencial que modela el amor entre Romeo y Julieta, se puede decir que nunca van a coincidir en un mismo sentimiento; es decir, cuando Julieta lo odia, el la ama y vicecersa, lo cual no es un amor sano pero es infinito.

SEGUNDO PUNTO

```python
# Definimos los parámetros
a = 2                                     
b = 1                                        
                                       
# Función f(t,y) que vamos a integrar
def armonico(y, t, a, b):
    #y1, y2 = y
    y1 = y[0]
    y2 = y[1]
    return [ (-a * y1) + (b * y2), (b * y1) +( -a * y2)]

y0 = [-2, 1]                                 
                                             

tt = np.linspace(0, 50, 500)                    

yy = odeint(armonico, y0, tt, args=(a, b))
```

```python
y1 = yy[:,0]
y2 = yy[:,1]
```

```python
label_size = 14
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size 

plt.figure(figsize=(8,6))
plt.plot(tt, y1, 'b', label='tiempo')
plt.plot(tt, y2, 'r', label='tiempo')
plt.legend(loc='best')
plt.xlabel('tiempo')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(y1, y2, c=y2, cmap='viridis')
plt.xlabel('tiempo')
plt.ylabel('tiempo')
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(12, 9))
ax = fig.gca(projection='3d')
ax.xaxis.set_pane_color((1,1,1,1))
ax.yaxis.set_pane_color((1,1,1,1))
ax.zaxis.set_pane_color((1,1,1,1))
ax.plot(tt, y1, y2, color='g', alpha=0.7, linewidth=1)

ax.set_title('El Amor Parte 2');
ax.set_xlabel('tiempo')
ax.set_ylabel('Romeo')
ax.set_zlabel('Julieta');

plt.show()
```

El amor de Romeo y Julieta es sano ya que cuando Julieta ama a Romeo, este tambien lo hace, existe reciprocidad. Conforme pasa el tiempo ambos se enamoran con la misma intensidad.
